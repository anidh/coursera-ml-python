```python
#Reading Data Using Numpy and Pandas
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
#The data we have extracted from the file
data = pd.read_csv('ex1data1.txt',names=['Population','Profits'])
data.head(n=10)
```

```python
#Describing The Data
data.describe()
```

```python
#Now Plotting The Data
data.plot(kind='scatter',x='Population',y='Profits',figsize=(12,8))
#Taking The X and The Y Values Here
X=data['Population'].values
originalValue=data['Profits'].values
print(X)
print(originalValue)
theta=0
alpha=0.001
```

```python
def trainingModel(x,original):
    global theta
    predictionValue=theta*x
    error=predictionValue-original
    lossFunction=(0.5)*(error**2)
    gradientDecent=alpha*error*x
    theta=theta-gradientDecent
    print(theta)
```

```python
iterations=10
print(X.size)
for x in range(iterations):
    for y in range(X.size):
        xValue=X[y]
        yValue=originalValue[y]
        trainingModel(xValue,yValue)
```

```python
from numpy import linspace as ls
line=ls(0,30,100)
plotLine=theta*line
mpl.plot(line,plotLine,'b')
mpl.plot(X,originalValue,'.')
```
