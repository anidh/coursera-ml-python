```python
#Reading our Data Here
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpl
%matplotlib inline
data= pd.read_csv('student.csv')
#Printing The Data
data.head(n=20)
```

```python
#Describinf The Data Here
data.describe()
```

```python
#Perfrominng Normalization On The Data
data=((data-data.mean())/data.std())
data.head()
```

```python
#Now Taking All The Individual Columns One By One
maths=data['Math'].values
reading=data['Reading'].values
write=data['Writing'].values
m=len(write)
X0=np.ones(m)
X=np.array([X0,maths,reading]).T
theta=np.array([0,0,0])
Y=np.array(write)
```

```python
#Now Performing Cost Computation
alpha=0.0001
iters=10000
def computeCost(X,Y,theta):
    inner=np.sum((X.dot(theta)-Y)**2)/(2*m)
    return inner
```

```python
computeCost(X,Y,theta)
```

```python
#Defining Gradient Descent
def gradientDescent(X,Y,alpha,iters,theta):
    costHist=[0]*iters
    for iterations in range(iters):
        #Defining The Hypothesis
        hypothesis=X.dot(theta)
        #Now Taking Out The Error
        error=hypothesis-Y
        gradient=X.T.dot(error)/m
        #Now Performing Gradient Descent Here
        theta=theta-alpha*gradient
        #Finding The New Cost Value
        newCost=computeCost(X,Y,theta)
        costHist[iterations]=newCost
    return theta,costHist
```

```python
newTheta,cost=gradientDescent(X,Y,alpha,10000,theta)
print(newTheta)
```
